# PROBLEM STATEMENTS
A retail company “ABC Private Limited” wants to understand the customer purchase behaviour (specifically, purchase amount) against various products of different categories. They have shared purchase summary of various customers for selected high volume products from last month. The data set also contains customer demographics (age, gender, marital status, city_type, stay_in_current_city), product details (product_id and product category) and Total purchase_amount from last month.

Now, they want to build a model to predict the purchase amount of customer against various products which will help them to create personalized offer for customers against different products.

```python
#importing the dataset
df_train = pd.read_csv("train.csv")
```

```python
##import test data
df_test = pd.read_csv("test.csv")
```

```python
df_train.head()
```

```python
#shape of the dataset
print(df_train.shape)
print(df_test.shape)
```

```python
## mearging the traing and test dataset
final_df = pd.concat([df_train,df_test])
```

```python
final_df.shape
```

```python
#info od the dataset
final_df.info()
```

```python
final_df.describe()
```

```python
#droping unnessary colums
final_df.drop(["User_ID"],axis=1,inplace=True)
```

```python
final_df.head(1)
```

```python
# handleling categorical feature GENDER
final_df['Gender'] = final_df['Gender'].map({'F':0 , 'M':1})
```

```python
#handle categorical feature AGE
final_df["Age"].unique()
```

```python
#ordinal encoading 
final_df["Age"] = final_df["Age"].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,
                                      '46-50':5,'51-55':6,'55+':7})
```

```python
#encoding city
final_df["City_Category"] = final_df['City_Category'].map({"A":1,"B":2,"C":3})
```

```python
#missing value
final_df.isnull().sum()
```

```python
####replacing null value
final_df['Product_Category_2'].unique()
```

```python
final_df['Product_Category_2'].value_counts()
```

```python
final_df['Product_Category_2'].mode()[0]
```

```python
#Replacing the null value with the mode
final_df['Product_Category_2'] = final_df['Product_Category_2'].fillna(final_df['Product_Category_2'].mode()[0])
```

```python
final_df['Product_Category_3'] = final_df['Product_Category_3'].fillna(final_df['Product_Category_3'].mode()[0])
```

```python
final_df.shape
```

```python
final_df['Stay_In_Current_City_Years'].unique()
```

```python
final_df['Stay_In_Current_City_Years']= final_df['Stay_In_Current_City_Years'].str.replace('+','')
```

```python
final_df.info()
```

```python
#convert object intoi integer
final_df['Stay_In_Current_City_Years']=final_df['Stay_In_Current_City_Years'].astype('int')
```
